```python
import plotly.express as px
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
col=['time','temp','event']
df = pd.read_csv("db_mod.csv")
df.columns=['time','temperature','event']
#df.head()
```

```python
temps = []
for temp in df.temperature:
    temps.append(temp)
#print(temps)
```

```python
temp_count = {}

for temp in temps:
    occ = temps.count(temp)
    temp_count.update({temp:occ})
tbl = pd.DataFrame.from_dict(temp_count, orient = "index")
tbl.columns = ['occurences']
#print(tbl)
```

```python
plt.figure(figsize=(12,6))
plt.title('Temperature occurences')
sns.lineplot(data=tbl)
```
